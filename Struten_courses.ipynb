Our 'Intermediate class Eligibility' column carries the boolean value that has to be treated and converrted into binary and as such , we would be creating a fucntion to conver the values to match our first dataset that carries (Yes/No) with the condition that the students is eligible only when the grade ponit average is either 50 or greater than 50

```python
def student(column):
    if column >=50 :
        return 'Yes'
    else:
        return 'No'
```

```python
train['Intermediate class Eligibility'] = train['Grade Point Average'].apply(student)

train
```

```python
## Coverting our Intermediate class Eligibility column to binary

train["Intermediate class Eligibility"].replace({'Yes' : 1,'No' : 0},inplace=True)
```

```python
train
```

```python
def fill_na_values(train: pd.DataFrame) -> pd.DataFrame:
    for features in train:
        if train[features].isna().sum() > 0:
            if train[features].dtype == 'object':
                train[features].fillna(value=train[features].mode()[0], inplace=True)
            else:
                train[features].fillna(value=train[features].median(), inplace=True)
    return train
```

```python
train=fill_na_values(train)
```

```python
# checking for the skewed cols
from scipy.stats import skew
skewed_cols = x_train.apply(lambda x: skew(x))
print(skewed_cols)
```

# it appears that our data does not have any significant skewness.

The output we got indicates that the skewness in our data is less than 0.5, which is usually considered as a small skewness. However, it is important to keep in mind that even a small degree of skewness can potentially affect the performance of some machine learning algorithms, such as linear regression. Therefore, it is generally a good practice to apply skewness correction techniques to our data, even if the skewness is small. This will help to ensure that our model is as accurate and generalizable as possible.

```python
# treating our skwed cols and creating a metric that would treat future skwed cols
from scipy.stats import boxcox

cols_to_transform = ['Quiz Summary', 'Assignment Summary', 'Grade Point Average']
transformed_data = x_train.copy() # make a copy of the original dataset

for col in cols_to_transform:
    if x_train[col].nunique() > 1: # skip constant columns
        transformed_data[col], _ = boxcox(x_train[col] + 1) # adding 1 to avoid zero values

# Check the skewness of the transformed variables
print(transformed_data[cols_to_transform].skew())
```

```python
from scipy.stats import boxcox

cols_to_transform = ['Quiz Summary', 'Assignment Summary', 'Grade Point Average']
transformed_test = x_test.copy() # make a copy of the original dataset

for col in cols_to_transform:
    if x_test[col].nunique() > 1: # skip constant columns
        transformed_test[col], _ = boxcox(x_test[col] + 1) # adding 1 to avoid zero values

# Check the skewness of the transformed variables
print(transformed_test[cols_to_transform].skew())
```
