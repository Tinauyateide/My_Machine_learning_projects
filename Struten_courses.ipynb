Our 'Intermediate class Eligibility' column carries the boolean value that has to be treated and converrted into binary and as such , we would be creating a fucntion to conver the values to match our first dataset that carries (Yes/No) with the condition that the students is eligible only when the grade ponit average is either 50 or greater than 50

```python
train
```

```python
def fill_na_values(train: pd.DataFrame) -> pd.DataFrame:
    for features in train:
        if train[features].isna().sum() > 0:
            if train[features].dtype == 'object':
                train[features].fillna(value=train[features].mode()[0], inplace=True)
            else:
                train[features].fillna(value=train[features].median(), inplace=True)
    return train
```

```python
train=fill_na_values(train)
```

## Exploring our column

To predict if a student is eligible for the next phase in the course based on the given dataset, some of the important columns to explore would be:

Grade Point Average (GPA): This column provides an overview of a student's overall performance throughout the course. Students who have consistently high GPAs are more likely to be eligible to graduate.

Intermediate class Eligibility: This column indicates whether a student is eligible to progress to the next level or not. If a student has consistently met the eligibility criteria for the intermediate class, then they are likely to be eligible to graduate the course.

Assignment Summary: The assignment summary column provides information about a student's performance on the assignments given throughout the course. Students who have consistently high scores on their assignments are more likely to be eligible to graduate.

Quiz Summary: The quiz summary column provides information about a student's performance on the quizzes given throughout the course. Students who have consistently high scores on their quizzes are more likely to be eligible to graduate.

By exploring these columns, we can build a model that predicts a student's eligibility to graduate the course with a reasonable degree of accuracy.

```python
gpa_column = train['Grade Point Average']

# Calculate the average quiz score
average_score = train['Assignment Summary'].mean()

# Identify the highest and lowest quiz scores
highest_score = gpa_column.max()
lowest_score = gpa_column.min()

# Count the number of students who scored above and below 50
num_students_above_50 = (gpa_column >= 50).sum()
num_students_below_50 = (gpa_column < 50).sum()

# Print the results
print("Average GPA Score:", average_score)
print("Highest GPA Score:", highest_score)
print("Lowest GPA Score:", lowest_score)
print("Number of Students Scoring Above 50:", num_students_above_50)
print("Number of Students Scoring Below 50:", num_students_below_50)

# Plotting the data as a bar chart
categories = ['Above 50', 'Below 50']
values = [num_students_above_50, num_students_below_50]

plt.bar(categories, values)
plt.xlabel('Score Category')
plt.ylabel('Number of Students')
plt.title('Students GPA')
plt.savefig('gpa1')
plt.show()

# Plotting the data as a pie chart
plt.figure(figsize=(10, 6))
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Percentage of Students GPA')
plt.savefig('gpa')
plt.show()
```

## Training the Models


```python
# classify function
from sklearn.model_selection import cross_val_score
def classify(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
    model.fit(x_train, y_train)
    print("Accuracy is", model.score(x_test, y_test)*100)
    # cross validation - it is used for better validation of model
    # eg: cv-5, train-4, test-1
    score = cross_val_score(model, x, y, cv=5)
    print("Cross validation is",np.mean(score)*100)
```

```python
from sklearn.linear_model import LogisticRegression
LR_model = model
model = LogisticRegression()
classify(model, x, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC_model = model
model = DecisionTreeClassifier()
classify(model, x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
RTC_model = model
model = RandomForestClassifier()
classify(model, x, y)
```
